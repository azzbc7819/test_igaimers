```python
#말그대로 범주형 데이터 > 수치형 데이터로 전환
qual_col = ['LINE', 'PRODUCT_CODE']

for i in qual_col:
    le = LabelEncoder()
    le = le.fit(x_data[i])
    x_data[i] = le.transform(x_data[i])


    for label in np.unique(df_test[i]): 
      if label not in le.classes_: 
        le.classes_ = np.append(le.classes_, label)
    df_test[i] = le.transform(df_test[i]) 
```

```python
from sklearn.impute import KNNImputer

KNN_imputer = KNNImputer(n_neighbors=10) #10개의 평균으로 계산

#test 데이터 nan 값 채우기기
filled_data = KNN_imputer.fit_transform(df_test)
df_test = pd.DataFrame(filled_data, columns=df_test.columns)
```

```python
df_test
```

```python

sep = {}
col_memo = {}
line = list(x_data['LINE'].unique())
p_code = list(x_data['PRODUCT_CODE'].unique())
columns = x_data.columns

for l in line:
    for p in p_code:
        sep[(l, p)] = pd.DataFrame(index=range(0), columns = columns)

for idx in range(len(x_data)):
    data = x_data.iloc[idx]
    data_l = data['LINE']
    data_p = data['PRODUCT_CODE']
    #print(data_l, data_p)
    sep[(data_l, data_p)] = pd.concat((sep[(data_l, data_p)], pd.DataFrame(x_data.iloc[idx]).transpose()), axis = 0)

new_sep = {}

for k in sep.keys():
    if len(sep[k]):
        sep[k].dropna(how = 'all', axis = 1, inplace = True)
        temp_col = list(sep[k].columns)
        sep[k][temp_col] = KNN_imputer.fit_transform(sep[k])
        temp_col.remove('Y_Class')
        temp_col.remove('LINE')
        temp_col.remove('PRODUCT_CODE')
        col_memo[k] = temp_col

        new_sep[k] = pd.DataFrame(sep[k])
        new_sep[k] = new_sep[k].drop(columns = ['LINE', 'PRODUCT_CODE'])

sep = new_sep

col_memo
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
for k in sep.keys():
  # 모델 학습
  model_rf = RandomForestClassifier(n_estimators=50, random_state=0)

  y_train = sep[k].pop('Y_Class').astype('int')
  x_train = sep[k].astype('float')

  model_rf.fit(x_train, y_train)
  sep[k] = model_rf

  print("훈련 세트 정확도 : {:.3f}".format(model_rf.score(x_train, y_train)))

result = []

for idx in range(len(df_test)):
  temp = df_test.iloc[idx]
  temp_l = temp.pop('LINE')
  temp_p = temp.pop('PRODUCT_CODE')
  temp = temp[col_memo[(temp_l, temp_p)]]
  temp = temp.values.reshape(1, -1)

  temp_rf = sep[(temp_l, temp_p)]
  y_pred = int(temp_rf.predict(temp))

  result.append(y_pred)
  
```
